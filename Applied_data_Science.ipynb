```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def read_data(file_name):
    lst=[]
    flag=0
    with open(file_name + '.csv', 'r',encoding='utf-8-sig') as file:
        data=file.readlines()
        for i in data:
            i=i.replace("\n","")
            i=i.replace("\ufeff","")
            i=i.replace('"','')
            lst.append(i.split(","))
            #print(lst)
    header=lst[0]
    lst= lst[1:]
    return lst,header

```

```python
baro,header= read_data('barometer-last-year')
baro = pd.DataFrame(baro)

baro.columns=header

baro
```

```python
baro['DateTime'] =  pd.to_datetime(baro['DateTime'],errors='coerce')

baro['Baro'] =  pd.to_numeric(baro['Baro'],errors='coerce')

baro
```

```python
rain,header = read_data('rainfall-last-year')

rain = pd.DataFrame(rain)

rain.columns= header
rain
```

```python
rain['DateTime'] =  pd.to_datetime(rain['DateTime'],errors='coerce')

rain['mm'] =  pd.to_numeric(rain['mm'],errors='coerce')

rain
```

```python
indoor,header = read_data('indoor-temperature-last-year')

indoor = pd.DataFrame(indoor)

indoor.columns= header
indoor
```

```python
indoor['DateTime'] =  pd.to_datetime(indoor['DateTime'],errors='coerce')

indoor['Humidity'] =  pd.to_numeric(indoor['Humidity'],errors='coerce')

indoor['Temperature'] =  pd.to_numeric(indoor['Temperature'],errors='coerce')

indoor['Temperature_range (low)'] =  pd.to_numeric(indoor['Temperature_range (low)'],errors='coerce')

indoor['Temperature_range (high)'] =  pd.to_numeric(indoor['Temperature_range (high)'],errors='coerce')

indoor
```

```python
outdoor,header = read_data('outside-temperature-last-year')

outdoor = pd.DataFrame(outdoor)

outdoor.columns= header
outdoor
```

```python
outdoor['DateTime'] =  pd.to_datetime(outdoor['DateTime'],errors='coerce')

outdoor['Temperature'] =  pd.to_numeric(outdoor['Temperature'],errors='coerce')

outdoor['Temperature_range (low)'] =  pd.to_numeric(outdoor['Temperature_range (low)'],errors='coerce')

outdoor['Temperature_range (high)'] =  pd.to_numeric(outdoor['Temperature_range (high)'],errors='coerce')

outdoor
```

```python
df['Month']=df['DateTime'].dt.month
df
```

```python
df['Season']= df["Month"].replace({3: "Spring",4: "Spring",5: "Spring", 6: "Summer",7: "Summer",8: "Summer",9: "Autumn",10: "Autumn",11: "Autumn",12: "Winter",1: "Winter",2: "Winter"})
df
```

```python
print('Temperature indoors vs outdoors per season')



plt.bar(df['Season'],df['Temperature_indoor'],width=0.8)
plt.bar(df['Season'],df['Temperature_outdoor'],width=0.8)
plt.legend(["Temperature_indoor" , "Temperature_outdoor"])
plt.show()
```

```python
#Temperateur indoors vs outdoors per month
import numpy as np
x=[]
y=[]
for i in range(1,13):
    temp=df[df['Month']==i]
    x.append(temp['Temperature_outdoor'].mean())
    y.append(temp['Temperature_indoor'].mean())

print(x)
print(y)

plt.plot(np.arange(1,13),x)
plt.plot(np.arange(1,13),y)
plt.legend(["Temperature_outdoor" , "Temperature_indoor"])
plt.show()

#put bar and line on this , use min and max of outdoor, bar use indoor outdoor
```

```python
aut=0
win=0
spr=0
su=0
for index,row in df.iterrows():
    if row['mm']!=0:
        if row['Season']=='Autumn':
                aut=aut+1
        elif row['Season']=='Winter':
                win=win+1
        elif row['Season']=='Spring':
                spr=spr+1
        elif row['Season']=='Summer':
                su=su+1
```

```python

y = np.array([aut, win, spr, su])
print(y)
mylabels = ["Autumn", "Winter", "Spring", "Summer"]
print('Rainy days per Season')
plt.pie(y, labels = mylabels,shadow=True,radius=3, autopct=lambda p : '{:.2f}%'.format(p,p * sum(y)/100))
plt.show() 
```
